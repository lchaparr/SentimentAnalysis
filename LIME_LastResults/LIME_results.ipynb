```python
final=list()
for i in range(len(feats)):
    ft=feats[i].split(",")
    a1=re.search('\(\'([a-z]*)', ft[0])
    b1=re.search('(.*)\)', ft[1])
    final.append((a1.group(1),b1.group(1)))
    print(a1.group(1))
```

```python
def Sort_Tuple(tup):  
      
    # getting length of list of tuples 
    lst = len(tup)  
    for i in range(0, lst):  
          
        for j in range(0, lst-i-1):  
            if (tup[j][1] > tup[j + 1][1]):  
                temp = tup[j]  
                tup[j]= tup[j + 1]  
                tup[j + 1]= temp  
    return tup 
```

```python
result=(Sort_Tuple(final))
print(result)
```
