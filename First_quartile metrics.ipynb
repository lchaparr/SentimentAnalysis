```python

def pval(metric1,metric2):
    p_val=scipy.stats.mannwhitneyu(metric1,metric2, use_continuity=True, alternative=None)
    return p_val.pvalue

def inequities005(pval1):
    return pval1<(0.05/28)

def inequities0005(pval1):
    return pval1<(0.005/28)

def ComparisonVals(metric1,metric2):
    value=pval(metric1,metric2)
    return (value,inequities005(value),inequities0005(value),metric1.name,metric2.name)
    
def pvalues(metric_BOW_SGD, metric_TFIDF_SGD, metric_BOW_BNB, metric_TFIDF_BNB, 
                  metric_BOW_LR, metric_TFIDF_LR,metric_BOW_MNB, metric_TFIDF_MNB):
    print(ComparisonVals(metric_BOW_SGD,metric_TFIDF_SGD))
    print(ComparisonVals(metric_BOW_SGD,metric_BOW_BNB))
    print(ComparisonVals(metric_BOW_SGD,metric_TFIDF_BNB))
    print(ComparisonVals(metric_BOW_SGD,metric_BOW_LR))
    print(ComparisonVals(metric_BOW_SGD,metric_TFIDF_LR))
    print(ComparisonVals(metric_BOW_SGD,metric_BOW_MNB))
    print(ComparisonVals(metric_BOW_SGD,metric_TFIDF_MNB))
    
    print(ComparisonVals(metric_TFIDF_SGD,metric_BOW_BNB))
    print(ComparisonVals(metric_TFIDF_SGD,metric_TFIDF_BNB))
    print(ComparisonVals(metric_TFIDF_SGD,metric_BOW_LR))
    print(ComparisonVals(metric_TFIDF_SGD,metric_TFIDF_LR))
    print(ComparisonVals(metric_TFIDF_SGD,metric_BOW_MNB))
    print(ComparisonVals(metric_TFIDF_SGD,metric_TFIDF_MNB))
    
    print(ComparisonVals(metric_BOW_BNB,metric_TFIDF_BNB))
    print(ComparisonVals(metric_BOW_BNB,metric_BOW_LR))
    print(ComparisonVals(metric_BOW_BNB,metric_TFIDF_LR))
    print(ComparisonVals(metric_BOW_BNB,metric_BOW_MNB))
    print(ComparisonVals(metric_BOW_BNB,metric_TFIDF_MNB))
    
    print(ComparisonVals(metric_TFIDF_BNB,metric_BOW_LR))
    print(ComparisonVals(metric_TFIDF_BNB,metric_TFIDF_LR))
    print(ComparisonVals(metric_TFIDF_BNB,metric_BOW_MNB))
    print(ComparisonVals(metric_TFIDF_BNB,metric_TFIDF_MNB))
    
    print(ComparisonVals(metric_BOW_LR,metric_TFIDF_LR))
    print(ComparisonVals(metric_BOW_LR,metric_BOW_MNB))
    print(ComparisonVals(metric_BOW_LR,metric_TFIDF_MNB))
    
    print(ComparisonVals(metric_TFIDF_LR,metric_BOW_MNB))
    print(ComparisonVals(metric_TFIDF_LR,metric_TFIDF_MNB))
    
    print(ComparisonVals(metric_BOW_MNB,metric_TFIDF_MNB))
```
